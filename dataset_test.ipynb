```python
import pandas as pd
import os

# กำหนด path
base_path = r'F:\Github Ai\hate-speech-dataset'
text_folder = os.path.join(base_path, 'sampled_test')
annotation_file = os.path.join(base_path, 'annotations_metadata.csv')

# โหลดไฟล์ annotations
df_ann = pd.read_csv(annotation_file)

# สร้างคอลัมน์ path ของไฟล์ .txt
df_ann['txt_path'] = df_ann['file_id'].astype(str) + '.txt'
df_ann['txt_path'] = df_ann['txt_path'].apply(lambda x: os.path.join(text_folder, x))

# อ่านเนื้อหาไฟล์ .txt ถ้ามีอยู่
def read_txt(path):
    try:
        with open(path, 'r', encoding='utf-8') as f:
            return f.read()
    except FileNotFoundError:
        return None  # หรือจะใส่เป็น '' ก็ได้

df_ann['text'] = df_ann['txt_path'].apply(read_txt)

# ตรวจสอบเฉพาะแถวที่อ่านไฟล์สำเร็จ
df_final = df_ann[df_ann['text'].notna()].copy()

# ตัวอย่างดูข้อมูล
df_final[['file_id', 'label', 'text']].head(10)
```
